# Guessing the distribution of `dataset 1-data_1`

```python
df = pd.read_excel('/kaggle/input/data-1/data_1.xlsx')
```

```python
df
```

```python
import matplotlib.pyplot as plt
plt.hist(df['obs'])
plt.show()
```

```python
df.describe()
```

```python
stats_labels = ['mean Observed', 'Variance Observed', 'Skew Observed', 'Kurt Observed']
stats_values = [round(df['obs'].mean(), 2), round(df['obs'].var(), 2), round(df['obs'].skew(), 2), round(df['obs'].kurt(), 2)]
print(set(zip(stats_labels, stats_values)))
```

```python
df.median()
```

```python
import statsmodels.api as sm
import scipy.stats as stats
```

```python
import matplotlib.pyplot as pltp
stats.probplot(df['obs'], plot=plt)
plt.show()
```

Using visualization, we got that distribution is poisson 

Now, using chi2 techninque, we will show that it is poisson

```python
print(df)
```

```python
# creating empty dataframe
grouped_df = pd.DataFrame()
```

```python
grouped_df['frequency'] = df.groupby('obs')['obs'].count()
```

```python
len(df)
```

```python
# now, we will calculate, observed probability = freq/ len(df)

grouped_df['Obs_Prob'] = grouped_df['frequency']/len(df)
print(grouped_df)
```

Now, we will calculate theoretical probability

```python
from scipy import stats
from scipy.stats import poisson 
from scipy.stats import chisquare
```

```python
poisson.pmf
```

```python
grouped_df.index
```

```python
df['obs'].mean()
```

- Obs_prob = in the sample, what is prob of that obs happening

- Poisson pmf = from population, with mean of 4.94, what is the probability of that obs happening 

```python
grouped_df['Poisson_pmf'] = poisson.pmf(k=grouped_df.index, mu=df['obs'].mean())
```

```python
grouped_df
```

```python
grouped_df['Poisson_Freq'] = round(grouped_df['Poisson_pmf']*len(df), 2)
```

```python
grouped_df
```

```python
obs_freq = list(grouped_df['frequency'])
expec_freq = list(grouped_df['Poisson_Freq'])
```

`NULL HYPOTHESIS`: The given data follows Poisson distribution
`ALTERNATE HYPOTHESIS`: The given data does not follow Poisson distribution

```python
obs_freq
```

```python
expec_freq
```

```python
chisquare(obs_freq, expec_freq)
```

```python
import scipy.stats
```

```python
scipy.stats.chi2.ppf(1-.05, df=9)
```

```python
stats_labels = ['Mean Observed', 'Variance Observed', 'Skew Observed', 'kurt Observed']
stats_values = [round(df['obs'].mean(), 2), round(df['obs'].var(), 2), round(df['obs'].skew(), 2), round(df['obs'].kurt(), 2)]
set(zip(stats_labels, stats_values))
```

- kurt is negative= that's why it is not gaussian.

- kurt is neg. means, we don't have enough evidence

```python
import statsmodels as sm
import scipy.stats as stats
from scipy.stats import norm, uniform, poisson
from statsmodels.graphics.gofplots import qqplot
```

```python
sm.graphics.gofplots.qqplot(df['obs'], stats.uniform, fit = True, line='45')
plt.show()
```

```python
import scipy.stats as stats
stats.probplot(df['obs'], dist='uniform', plot=plt)
plt.show()
```

since, it's straight line in both ppplot and qqplot, that's why it is uniform dist.
